```python
from parsel import Selector

with open("../storage/latest_share_price_scroll_l-20231018.html", encoding="utf8") as content:
    html_sel = Selector(content.read())

table_sel = html_sel.css("div.table-responsive.inner-scroll")
table_wrapper = table_sel.get()

# remove all extra </tbody>
table_wrapper = table_wrapper.replace("</tbody>", "").replace("</table>", "</tbody></table>")

content_sel = Selector(table_wrapper)

header_sel = content_sel.css("thead > tr > th::text").getall()
print(header_sel)

body_sel = content_sel.css("tbody > tr")

def prepare_cell_data(item: str):
    if "displayCompany" in item:
        idx1 = item.index("?name=") + len("?name=")
        item = item[idx1:]

        idx2 = item.index('"')
        return item[:idx2]
    else:
        content_sel = Selector(item)
        return content_sel.css("td::text").get()

for tablerow in body_sel:
    table_cells = tablerow.css("tr > td").getall()
    print(list(map(prepare_cell_data, table_cells)))
```
